# 將Bitstream File 燒錄至板子上

```python
from pynq import Overlay
Lab_Design = Overlay("./lab6.bit")
```

# 從硬體檔案中讀取這個系統中的IP資訊

```python
Lab_Design.ip_dict
```

# 將IP的Address從ip_dict中提出來

```python
BRAM0_ADDR = Lab_Design.ip_dict['axi_bram_ctrl_0']['phys_addr']
BRAM0_ADDR_range = Lab_Design.ip_dict['axi_bram_ctrl_0']['addr_range']
GPIO0_ADDR = Lab_Design.ip_dict['axi_gpio_0']['phys_addr']
GPIO0_ADDR_range = Lab_Design.ip_dict['axi_gpio_0']['addr_range']
GPIO1_ADDR = Lab_Design.ip_dict['axi_gpio_1']['phys_addr']
GPIO1_ADDR_range = Lab_Design.ip_dict['axi_gpio_1']['addr_range']
GPIO2_ADDR = Lab_Design.ip_dict['axi_gpio_2']['phys_addr']
GPIO2_ADDR_range = Lab_Design.ip_dict['axi_gpio_2']['addr_range']
print("The Base Address of BRAM0 is : " + hex(BRAM0_ADDR))
print("The Address Range of BRAM0 is : " + hex(BRAM0_ADDR_range))
print("The Base Address of GPIO0 is : " + hex(GPIO0_ADDR))
print("The Address Range of GPIO0 is : " + hex(GPIO0_ADDR_range))
print("The Base Address of GPIO1 is : " + hex(GPIO1_ADDR))
print("The Address Range of GPIO1 is : " + hex(GPIO1_ADDR_range))
print("The Base Address of GPIO2 is : " + hex(GPIO2_ADDR))
print("The Address Range of GPIO2 is : " + hex(GPIO2_ADDR_range))
```

# 使用Pynq 內建的 MMIO (Memory-mapped I/O) 以達成 PS 和 AXI IP 之間的溝通
## MMIO(IP_BASE_ADDRESS, ADDRESS_RANGE)

[MMIO Document](https://pynq.readthedocs.io/en/v2.7.0/pynq_libraries/mmio.html)

```python
from pynq import MMIO
BRAM0_MMIO = MMIO(BRAM0_ADDR, BRAM0_ADDR_range)
GPIO0_MMIO = MMIO(GPIO0_ADDR, GPIO0_ADDR_range)
GPIO1_MMIO = MMIO(GPIO1_ADDR, GPIO1_ADDR_range)
GPIO2_MMIO = MMIO(GPIO2_ADDR, GPIO2_ADDR_range)
```

# 以下程式行為皆和Lab5-2的main.c相同
## 只是改以Python形式撰寫

```python
a = 0x00000010
addr_cmd = 0b00000000010100001000110000000000
print("Lab 5-2 Program Start.")
```

# mmio.write(ADDRESS_OFFSET, data)

```python
BRAM0_MMIO.write(0, a)
GPIO0_MMIO.write(0, 0x0000003A)
GPIO1_MMIO.write(0, addr_cmd)
```

# mmio.read(ADDRESS_OFFSET)

```python
a_read = GPIO2_MMIO.read(0)
print("Data = ", hex(a_read))
```
